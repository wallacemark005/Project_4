```python
# Libraries
import sys
import scipy
import numpy as np
import matplotlib
import pandas as pd
import sklearn
import os

from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(os.path.join("Resources", "country_population.csv"))

deu=df.loc[df['Country Name']=='Germany']
deu.drop(['Country Name','Country Code','Indicator Name','Indicator Code'],axis=1,inplace=True)
deu = deu.T

deu.dropna(inplace=True)
deu=deu.reset_index().rename(columns={53:'fr','index':'year'})

x = deu.iloc[:, 0].values.reshape(-1, 1)
y = deu.iloc[:, 1].values.reshape(-1, 1)
model = LinearRegression().fit(x, y)
y_pred = model.predict([[1964]])
y_pred
```
