```python
# Libraries
import sys
import scipy
import numpy as np
import matplotlib
import pandas as pd
import sklearn
import os

from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(os.path.join("Resources", "country_population.csv"))
df.head()
```

```python
sgp=df.loc[df['Country Name']=='Singapore']
sgp.drop(['Country Name','Country Code','Indicator Name','Indicator Code'],axis=1,inplace=True)
sgp = sgp.T
sgp.head()

sgp.dropna(inplace=True)
sgp=sgp.reset_index().rename(columns={18:'population','index':'year'})
sgp.head()

x = sgp.iloc[:, 0].values.reshape(-1, 1)
y = sgp.iloc[:, 1].values.reshape(-1, 1)
model = LinearRegression().fit(x, y)
y_pred = model.predict([[2055]])
y_pred
```
